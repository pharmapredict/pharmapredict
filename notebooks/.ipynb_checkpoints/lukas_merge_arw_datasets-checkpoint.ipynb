# Merging the authorised/rejected with the withdrawn dataset

```python
import pandas as pd
import numpy as np
```

```python
df_ar = pd.read_excel('../raw_data/20201119/Medicines_output_european_public_assessment_reports_20201119.xlsx')
df_w = pd.read_excel('../raw_data/20201119/Medicines_output_withdrawn_applications_20201119.xlsx')
```

```python
df_ar.head()
```

```python
df_w.head()
```

## Filter dataframes

```python
df_ar['Authorisation status'].unique()
```

```python
df_ar = df_ar[df_ar['Category'] == 'Human']
df_ar = df_ar[(df_ar['Authorisation status'] == 'Authorised') | (df_ar['Authorisation status'] == 'Refused')]
```

```python
df_w = df_w[df_w['Category'] == 'Human']
df_w = df_w[df_w['commercial reason'] == 'n']
df_w = df_w[df_w['Type of withdrawal'] == 'Initial authorisation']
```

## Select relevant columns

```python
df_ar.columns
```

```python
df_ar = df_ar[['Category', 'Medicine name', 'Therapeutic area',
       'INN or active substance', 'Authorisation status', 
       'Orphan medicine',
       'Marketing authorisation holder/company name',
       'earliest date']]
```

```python
df_w.columns
```

```python
df_w['Authorisation status'] = 1
```

```python
df_w = df_w[['Category', 'Medicine name', 'Therapeutic area',
       'INN or active substance', 'Authorisation status', 
       'Orphan medicine',
       'Marketing authorisation holder/company name',
       'earliest date']]
```

```python
df_w.columns == df_ar.columns
```

## Merge dataframes

```python
df = pd.concat([df_ar, df_w], ignore_index=True)
```

```python
df
```

```python
df[df.duplicated(subset='INN or active substance')].sort_values('INN or active substance')
```

```python
# replace []
df['Therapeutic area'] = df['Therapeutic area'].replace("[", "").replace("]", "")
df['INN or active substance'] = df['INN or active substance'].replace("[", "").replace("]", "")
```

```python
df['Authorisation status'].unique()
```

```python
# map authorisation status
df['Authorisation status'] = df['Authorisation status'].map({'Authorised': 0,
                                                             'Refused': 1,
                                                             1: 1})
```

```python
# map Orphan medicine
df['Orphan medicine'] = df['Orphan medicine'].map({'no': 0, 'yes': 1})
```

```python
df
```

```python
df[df['Authorisation status'] == 1]
```

```python
df.to_csv('../raw_data/arw.csv')
```
