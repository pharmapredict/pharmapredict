# Save last n sentences of abstracts to dataset

```python
import pandas as pd
import numpy as np
```

```python
# base dataframe
df = pd.read_csv('../raw_data/enriched_CT.csv')
```

```python
df
```

```python
from nltk.tokenize import sent_tokenize
```

```python
def n_last_senteces(text, n):
    return ' '.join(sent_tokenize(text)[-n:])
```

```python
df["conclusions"] = ""
```

```python
df_abstr = pd.read_csv('abstracts_by_id.csv')
```

```python
for index, row in df.iterrows():
    if index < 100_000:
        print(index)
        conclusions = ""
        for _, row_abstr in df_abstr.loc[df_abstr['id'] == index].iterrows():
            conclusion = n_last_senteces(row_abstr['abstract'], 2)
            conclusions += " " + conclusion
        df['conclusions'][index] = conclusions    
```
